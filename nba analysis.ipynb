```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nba.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print((df.isnull().mean() * 100).round(2))
```

```python
df.drop(columns=df.columns[df.isnull().mean() > 0.4], inplace=True)
```

```python
categorical_col = df.select_dtypes(include= ([np.number])).columns
for col in categorical_col :
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
text_columns = df.select_dtypes(include= 'object').columns
```

```python
for col in text_columns :
    df[col]= df[col].fillna("unknown")
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.values
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Player', order=df['Player'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribution of Players")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = df, x = 'Player', order = df['Player'].value_counts().head(10).index)
plt.title("count of players ")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
           
```
